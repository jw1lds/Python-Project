```python
#plots top living wage of top 10 US cities by that metric as bar graph
data_source_file = 'livingwage.csv'

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv(data_source_file)
wage = data.one_adult_no_kids_living_wage
city = data.city

#sorts cities by solo adult w. no kids livable wage
wageordered = wage.sort_values(ascending=True)

#plotting parameters
plt.title('Living Wage in Top 10 US Cities')
plt.xlabel('City')
plt.ylabel('Living Wage ($/hr)')
plt.xlim([0, 9])
plt.bar(city, wageordered)
```
